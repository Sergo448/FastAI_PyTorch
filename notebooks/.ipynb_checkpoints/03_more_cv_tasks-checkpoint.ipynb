```python
from fastai.vision.all import *
import torch
import pandas as pd
```

```python
import torchvision

torch.cuda.is_available()
```

```python
path = untar_data(URLs.PASCAL_2007)
```

```python
df = pd.read_csv(path/'train.csv')
df.head()
```

```python
# Блок данных без параметров
dblock = DataBlock()
# на основе datasets
dsets = dblock.datasets(df)
```

```python
dsets.train[0]
```

```python
# пример без lambda func
def gets_x(r): return path/'train'/r['fname']
def gets_y(r): return r['labels'].split(' ')
# dblock = DataBlock(get_x=gets_x, get_y=gets_y) etc.

# get_x & get_y необъодимы для выделения необходимых полей
dblock = DataBlock(get_x = lambda r: r['fname'], get_y = lambda r: r['labels'])
dsets = dblock.datasets(df)
dsets.train[0]
```

```python
dblock = DataBlock(get_x=gets_x, get_y=gets_y)
dsets = dblock.datasets(df)
dsets.train[0]
```

```python
dblock = DataBlock(blocks=(ImageBlock, MultiCategoryBlock),
                   get_x=gets_x, get_y=gets_y)
dsets = dblock.datasets(df)
dsets.train[0]
```

```python
idxs = torch.where(dsets.train[0][1]==1.)[0]
dsets.train.vocab[idxs]
```

```python
def splitter(df):
    train = df.index[~df['is_valid']].tolist()
    valid = df.index[df['is_valid']].tolist()
    return train, valid
```

```python
dblock = DataBlock(blocks=(ImageBlock, MultiCategoryBlock),
                   splitter=splitter,
                   get_x=gets_x, get_y=gets_y)
dsets = dblock.datasets(df)
dsets.train[0]
```

```python
dblock = DataBlock(blocks=(ImageBlock, MultiCategoryBlock),
                   splitter=splitter,
                   get_x=gets_x, get_y=gets_y,
                   item_tfms=RandomResizedCrop(128, min_scale=0.35))
dls = dblock.dataloaders(df)
```

```python
dls.show_batch(nrows=3, ncols=3)
```

```python
learn = vision_learner(dls, resnet18).cuda()
```

```python
"""model = torch.load(PATH).type(torch.FloatTensor).to(device)
input = input.type(torch.FloatTensor).to(device)
"""
# model = model.to("cuda")
# data = data.to("cuda")

x, y = dls.train.one_batch()


activs = learn(x)
activs.shape
```

```python
activs[0]
```

```python
def binary_cross_entropy(inputs, targets):
    inputs = inputs.sigmoid()
    return -torch.where(targets==1, inputs, 1-inputs).log().mean()
```

```python
loss_func = nn.BCEWithLogitsLoss()
loss = loss_func(activs, y)
loss
```

```python
def accuracy_multi(inp, targ, thresh=0.5, sigmoid=True):
    "Compute accuracy when 'inp' and 'targ' are the same size."
    if sigmoid: inp = inp.sigmoid()
    return ((inp>thresh) == targ.bool()).float().mean()
```

```python
learn = vision_learner(dls, resnet50, metrics=partial(accuracy_multi, thresh=0.2)).cuda()
learn.fine_tune(3, base_lr=3e-3, freeze_epochs=4)
```
